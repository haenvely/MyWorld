```python
import cx_Oracle  # 특이하게도 Oracle이란 것....
```

```python
# cx_oracle이 갖고 있는 connect를 사용하면 됨
# 내 ip = localhost = 127.0.0.1
conn = cx_Oracle.connect('hr', 'hr', '127.0.0.1:1521/xe')  # ID, PWD, 서버 주소: 포트 번호/아이디(보통 xe). ipconfig
conn
```

```python
query = 'select * from departments'
cursor = conn.cursor() #객체 생성
cursor.execute(query)

cursor
```

```python
for row in cursor:
    print(row)
```

```python
# 잘 Close 해주기 #
cursor.close()
conn.close()
```

```python
conn = cx_Oracle.connect('scott', 'tiger', '127.0.0.1:1521/xe')  # ID, PWD, 서버 주소: 포트 번호/아이디(보통 xe). ipconfig

query = '''
insert into cxo_test(id, name) values(1, 'kang')
'''
cursor = conn.cursor()
cursor.execute(query)
conn.commit() ## commit, rollback 해야 트랜잭션이 끝남.

cursor.close()
conn.close()
```

```python
conn = cx_Oracle.connect('scott', 'tiger', '127.0.0.1:1521/xe')  # ID, PWD, 서버 주소: 포트 번호/아이디(보통 xe). ipconfig

query = '''
insert into cxo_test(id, name) values(:1, :2)
'''
cursor = conn.cursor()
cursor.execute(query, (2, '고양시'))
conn.commit() ## commit, rollback 해야 트랜잭션이 끝남.

cursor.close()
conn.close()
```
